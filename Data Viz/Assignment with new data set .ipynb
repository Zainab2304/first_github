```python
#importing libraries 
import seaborn as sns 
import matplotlib.pyplot as plt 
import pandas as pd
import numpy as np
#importing data set 
my_data= pd.read_csv("new_data.csv")
#setting canvas 
sns.set(style="darkgrid")
#setting final figure size 
plt.figure(figsize=(18,12))


sns.boxplot(x="Qualification_completed", y="Your Weight in kg? (float)" ,hue="Gender", dodge= True ,data= my_data, saturation= 1, showmeans=True, meanprops={"marker":"+", "markersize":"15", "markeredgecolor":"red"})
plt.title("Kon Ziada Khata hai ?", size= 28, weight= "bold")
plt.show()
```
